```python
import numpy as np
import random
import matplotlib.pyplot as plt
m=[1, -1]
n=int(input('Введите размер матрицы '))
a=np.zeros((n,n))
h=int(input('Введите количество повторов '))
l=0 
per=int(input('Нешахматы(1) или шахматы(2)'))   #Заполнение матрицы 1/-1 или чередованием(шахматный порядок)
for i in range(n):
    for j in range(n):
        a[i, j]=random.choice(m)
print(a)                                   
c=a.copy()

for l in range(h):                              #цикл для заполнения матрицы
    print("Петля номер", l)
    b=a.copy()
    summ1=0
    summ2=0
    b[random.randrange(n), random.randrange(n)]*=-1 
    for i in range(n):
        for j in range(n):
            summ1+=a[i, j]*a[i, j-1]+a[i-1, j]*a[i, j]
            summ2+=b[i, j]*b[i, j-1]+b[i-1, j]*b[i, j]
    print('Сумма новой ',summ2)
    if per==1:
        dsumm=-1*(summ2-summ1)
    else:
        dsumm=summ2-summ1
    print('Разницы сумм ',dsumm)                                   
    if dsumm<=0:                                       
        a=b.copy()
    else:
        t=float(1)
        w=np.exp(-dsumm/t)
        p=round(random.uniform(0, 1), 5)
        if w>=p:
            c=a.copy()
        elif w<0:
            a=c.copy()
    i+=1
    plt.clf()                                  #графический вид получившейся в конце цикла матрицы
    plt.imshow(a)
    plt.draw()
    plt.show()
    plt.gcf().canvas.flush_events()
```

##### 1.генерируем стартовую случайную матрицу, вычисляем Е1(наша сумма попарных произведений соседей)
2.Выбираем случайным образом элемент матрицы и меняем его знак на противоположный.
3.Вычисляем ∆Е, то есть изменение энергии системы.
4.Если ∆Е меньше или равно нулю, то принимаем новую конфигурацию и переходим к шагу 2.
5.Если ∆Е положительно, вычисляем «вероятность перехода»:
-Генерируем случайное число P в интервале (0, 1)
-вычисляем по формуле величину W = -∆Е / T, где T- константа которую задаем где-нибудь в начале кода. для начала пусть T = 0.8
-Если P≤W, то новую конфигурацию принимаем, переходим к шагу 2
-P>W то конфигурацию не принимаем(т.е. возвращаем предыдущую), переходим к шагу 2

Повторяем шаги 2–5, начать можно с матрицы 6х6 на 300 итераций. интерес представляет конечная конфигурация и ее значение Е. Также можно с помощью plt.imshow() вывести на экран изображение
