```python
import random
class rpg():
    global n
    n = 0
    def __init__(self, name, race, damage=0, potion=0, hp=0, stamina=0, gold=5 , maxhp=40, maxstamina=3):
        maxstamina = 3
        defdam = 10
        defhp = 40
        defheal = 3
        self.name = name
        self.race = race
        self.damage = defdam
        self.hp = defhp
        self.potion = defheal
        self.stamina = maxstamina
        self.gold = gold
        self.maxhp = defhp
        self.maxstamina = 3
    def dunge(self):
        if self.stamina > 0:
            d=int(round(random.random(), 1))
            if d>=0.2:
                self.stamina -= 1
                self.gold += 10
                print('You came out of the dungeon with severe fatigue and shortness of breath and found 10 gold coins in a chest near the exit(gold: +10)')
                print('Now you have '+str(self.gold)+' gold', ' stamina: -1, current stamina level: '+str(self.stamina))
            else:
                self.stamina -= 1
                print('Right before the exit you were hit by the arrow fired from the trap you triggered second before(hp: -10)')
                print('After thic accident you see a chest in front of you(gold: +15)')
                self.hp -= 10
                self.gold += 15
                print('Now you have '+str(self.gold)+' gold', ' stamina: -1, current stamina level: '+str(self.stamina))
        else:
            print('You need to have some rest')
        
    
    def rest(self):
        c = int(round(random.random(), 0))
        if self.stamina < self.maxstamina:
            if c == 0:
                self.hp -= 2
                self.stamina += 1
                print('You have slept for a while...')
                print("but still sleeping on the ground wasn't the best idea")
                print('*you feel terrible back pain* (hp -2)')
                print('hp: '+str(self.hp), 'stamina: '+str(self.stamina))
            else:
                print('You have slept for a while, you were woken up by the rays of the morning sun')
        else:
            print("You don't need a rest right now")
        
    def tavern(self):
        if self.gold < 6:
            print('Not enough gold, pal')
        else:
            self.gold -= 6
            if self.stamina == self.maxstamina:
                print('Ooops')
                print('Your stamina has been at its maximum all this time')
                print('You have just lost 6 gold :(')
            else:
                self.stamina = self.maxstamina
                print('You feel rested now')
    
    def Gymboss(self):
        global n
        if self.stamina > 0:
            if n!=1:
                print('You are attacking the boss')
                k = round(random.random(), 1)
                    
                if k == 0.7 or k == 0.8:
                    print('Your luck has failed you more than once, this time was no exception, but it definitely became the most memorable')
                    print("You stumble and fall on your own sword, that's a shame, it seemed even the boss was looking at you with a surprised grimace on his face")
                    self.hp -= self.damage
                    print('Your hp: '+str(self.hp))
                    
                elif k == 0.9:
                    self.boss_hp -= 1.5 * self.damage
                    print("You dodge the boss's attack and hit him right in his ugly head")
                    print("Boss's hp:" +str(self.boss_hp))
                    
                elif k == 1:
                    print('It was critical and may the last mistake in your god damn life')
                    self.hp -= 2 * self.boss_damage
                    print('Your hp: '+str(self.hp))
                else:  
                    self.hp -= self.boss_damage
                    self.boss_hp -= self.damage
                    print('Your hp: '+str(self.hp))
                    print("Boss's hp: "+str(self.boss_hp))
                self.stamina -= 1
                if self.hp <= 0:
                    print('You just died and fed the boss yourself')
                    self.boss_maxhp += self.maxhp
                    self.boss_maxdamage += round(self.damage/2, 0)
                    self.maxhp = 30
                    self.hp = self.maxhp
                    self.damage = 10
                    self.gold = 2
                    self.stamina = 3
                    self.potion = 3
                        
                    maxstamina = 3
                    self.boss_hp = self.boss_maxhp
                    self.boss_damage = self.boss_maxdamage
                    print("Bos's stats: hp - "+str(self.boss_hp), 'damage - '+str(self.boss_damage))
                    print("Now you will reborn in your original form")
                    print('The characteristics have been reset to zero')
                else:
                    print('Well done')     
                if self.boss_hp <= 0:
                    print('Boss was killed by '+str(self.name))
                    self.damage += 5
                    self.hp += 20
                    self.gold += 30
                    n=1
            else:
                print('Boss is already dead')
        else:
            print('You need to have some rest')
    
    def heal(self):
        if self.hp != self.maxhp and self.potion > 0:
            if self.maxhp - self.hp >= 20:
                self.potion -= 1
                self.hp += 20
                print('You are using a heal potion')
                print('current hp: '+str(self.hp))
                print('now you have '+str(self.potion)+' potions')
            else:
                self.potion -= 1
                self.hp = self.maxhp
                print('You are using a heal potion')
                print('current hp: '+str(self.maxhp))
                print('now you have '+str(self.potion)+' potions')
        else:
            print("you can't use it right now")
    
    def shop(self):
        if 'elf' in self.race or 'Elf' in self.race:
            print('Get out of here dirty elf')
            print('We have nothing to do with elves')
        elif 'orc' in self.race or 'Orc' in self.race:
            print('input')
            print('1 to buy heal potion for 8 gold')
            print('2 to increase your damage by 2 for 5 gold')
            print('3 to increase your max hp by 4 for 5 gold')
            print('4 to increase your max stamina level by 1 for 9 gold')
            k=input('input 1/2/3/4 ')
            if k==1 and self.gold >= 8:
                self.gold -= 8
                self.potion += 1
            elif k==2 and self.gold >= 5:
                self.gold -= 5
                self.damage += 2
            elif k==3 and self.gold >= 5:
                self.gold -= 5
                self.hp += 4
                self.maxhp += 4
            elif k==4 and self.gold >= 9:
                self.gold -=9
                self.maxstamina += 1
            else:
                if k!=1 and k!=2 and k!=3 and k!=4:
                    print("It's illigal answer")
                else:
                    print('You have not enough gold')
        else:
            print('input')
            print('1 to buy heal potion for 6 gold')
            print('2 to increase your damage by 2 for 3 gold')
            print('3 to increase your max hp by 4 for 3 gold')
            print('4 to increase your max stamina level by 1 for 7 gold')
            k=input('input 1/2/3/4 ')
            if k==1 and self.gold >= 6:
                self.gold -= 6
                self.potion += 1
                print("You've bought a potion")
                print('Now you have '+str(self.potion))
            elif k==2 and self.gold >= 3:
                self.gold -= 3
                self.damage += 2
                print("You've improved the sword")
                print('Damage: '+str(self.damage))
            elif k==3 and self.gold >= 3:
                self.gold -= 3
                self.hp += 4
                self.maxhp += 4
                print("You've bought new armor")
                print('current hp: '+str(self.hp))
                print('max hp: '+str(self.maxhp))
            elif k==4 and self.gold >= 7:
                self.gold -= 7
                self.maxstamina += 1
                print("You've got strange essence")
                print('current stamina: '+str(self.stamina))
                print('max stamina: '+str(self.maxstamina))
            else:
                if k!=1 and k!=2 and k!=3 and k!=4:
                    print("It's illigal answer")
                else:
                    print('You have not enough gold')
    def stats(self):
        print('name: '+self.name) 
        print('race: '+str(self.race))
        print('gold: '+str(self.gold))
        print('damage: '+str(self.damage)) 
        print('maxhp: '+str(self.maxhp)) 
        print('current hp: '+str(self.hp)) 
        print('max stamina level: '+str(self.maxstamina))
        print('current stamina level: '+str(self.stamina)) 
        print('potions: '+str(self.potion)) 
    
```
