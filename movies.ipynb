```python
#Таблица пользователей
import pandas as pd
vnames = ['UserID', 'gender', 'age', 'Occupation', 'Zip-code']
users = pd.read_table('C:\\Users\\artem\Desktop\\soldatov_ks-data-69c734973a6a\\movielens\\users.dat',
                      sep='::', engine='python', names=vnames)
users 
```

```python
#Таблица рейтингов
rate = ['UserID', 'MovieID', 'Rating', 'Timestamp']
ratings = pd.read_table('C:\\Users\\artem\\Desktop\\soldatov_ks-data-69c734973a6a\\movielens\\ratings.dat',
                        sep='::', engine='python', names=rate)
ratings
```

```python
#Таблица фильмов
films = ['MovieID', 'Title', 'Jenre']
movies = pd.read_table('C:\\Users\\artem\\Desktop\\soldatov_ks-data-69c734973a6a\\movielens\\movies.dat',
                        sep='::', engine='python', names=films)
movies
```

```python
#Объединение рейтингов и пользователей
urate = pd.merge(ratings, users)
urate
```

```python
#Объединение всего в одну
kpsk = pd.merge(urate, movies)
kpsk
```

```python
#СР рейтинг по полу
srate = kpsk.pivot_table('Rating', index='Title', columns='gender', aggfunc = 'mean')
srate
```

```python
#СР рейтинг по возрасту
sage = kpsk.pivot_table('Rating', index='Title', columns='age', aggfunc = 'mean')
sage
```

```python
#Сортировка по женским оценкам
resultg = srate.sort_values('F', axis=0, ascending=False)
resultg
```

```python
#Сортировка по возрастным оценкам
resulta = sage.sort_values(56, axis=0, ascending=True)
resulta
```

```python
#Сортировка по количеству отзывов
r_b_title = kpsk.groupby('Title')
l = r_b_title.size () 
resultR = l.sort_values()
resultR[resultR >= 250]
```

```python
#Разница в отзывах
srate = kpsk.pivot_table('Rating', index='Title', columns='gender', aggfunc = 'mean')
DIF=srate['F'] - srate['M'] 
resultDIF = DIF
DIF[DIF >= 0]
```

```python
#Женские - высокие, мужские - низкие
srate = kpsk.pivot_table('Rating', index='Title', columns='gender', aggfunc = 'mean')
DIF=srate['F'] - srate['M'] 
resultDIF = DIF.sort_values(ascending=False)
k = resultDIF[resultDIF >=0]
m = k.head(15)
m
```

```python
#Мужские - высокие, женские - низкие
srate = kpsk.pivot_table('Rating', index='Title', columns='gender', aggfunc = 'mean')
DIF=srate['M'] - srate['F'] 
resultDIF = DIF.sort_values(ascending=False)
k = resultDIF[resultDIF >= 0]
m = k.head(15)
m
```
